```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class SVM_simpleSMO():
    def loadDataSet(self, filename):
        """
        Desc:
            load dataset from .txt file.
        Args:
            filename -- the .txt file path
        Returns:
            dataMat -- feature matrix
            labelMat == sample labels
        """
        dataMat = []
        labelMat = []
        with open(filename) as fr:
            for line in fr.readlines():
                lineArr = line.strip().split('\t') # the element is string type
                dataMat.append([float(lineArr[0]), float(lineArr[1])]) # why it didn't add x0?
                labelMat.append(float(lineArr[2]))
        return dataMat, labelMat
    
    def selectJrand(self, i, m):
        """
        Desc:
            select a subscript of alpha randomly with a giving i (j ≠ i)
        Args:
            i -- the first subscript of alpha we selected
            m -- the total number of alpha subscript
        Returns:
            j -- the second alpha subscript
        """
        l = list(range(m))
        seq = l[: i] + l[i+1:]
        return np.random.choice(seq) # np.random.choice(seq): Generates a random sample from a given 1-D array
        
    def clipAlpha(self, aj, H, L):
        """
        Desc:
            clip the aj after update.
            More details can be found in https://zhuanlan.zhihu.com/p/29212107.
        Args:
            aj -- aj after update
            H -- high limit of aj
            L -- low limit of aj
        Return:
            aj -- aj after clipping
        """
        if aj > H:
            return H
        elif aj < L:
            return L
        else:
            return aj
        
    def smoSimple(self, dataMatIn, classLabels, C, toler, maxIter):
        """
        Desc:
            smo algorithm. No heuristic algorithm is used to find the second alpha subscript to maximum |Ei-Ej|.
        Args:
            dataMatIn -- feature matrix
            classLabels -- sample labels
            C -- variable of relaxing
            toler -- margin of error
            maxIter -- max iteration time
        Returns:
            b -- constant in f(x)
            alphas -- lagrangian multiplier in lagrangian function
        """
        dataMatrix = np.mat(dataMatIn)
        labelMat = np.mat(classLabels).transpose()
        m, n = np.shape(dataMatrix)
        
        b = 0
        alphas = np.mat(np.zeros((m, 1)))
        
        iter = 0 # alphas没有任何改变时，连续遍历X的次数
        while (iter < maxIter):
            alphaPairsChanged = 0 # flag value, it will be set as 0 at the beginning of each iteration step.
            for i in range(m):
                # f(xi) = wTxi + b, w=Σ(j:1~n)(alpha[j]*label[j]*x[j])
                fXi = float(np.multiply(alphas, labelMat).T * (dataMatrix * dataMatrix[i, :].T)) + b
                # error
                Ei = fXi - float(labelMat[i])
                # only if labelMat[i]*Ei exceeds the tolerance, should we perform the optimization on alpha pairs
                if ((labelMat[i] * Ei < -toler) and (alphas[i] < C)) or ((labelMat[i] * Ei > toler) and (alphas[i] > 0)):
                    j = self.selectJrand(i, m)
                    fXj = float(np.multiply(alphas, labelMat).T * (dataMatrix * dataMatrix[j, :].T)) + b
                    Ej = fXj - float(labelMat[j])
                    alphaIold = alphas[i].copy()
                    alphaJold = alphas[j].copy()
                    
                    # use L,H to limit new alpha[j]
                    if (labelMat[i] != labelMat[j]): 
                        L = max(0, alphas[j] - alphas[i])
                        H = min(C, C + alphas[j] - alphas[i])
                    else:
                        L = max(0, alphas[j] + alphas[i] - C)
                        H = min(C, alphas[j] + alphas[i])
                    if L == H: # in this case we can't optimize alphas
                        print ("L = H")
                        continue
                    
                    # eta参考《统计学习方法》P125~P128, attention that the eta is opposite to 
                    # that in https://zhuanlan.zhihu.com/p/29212107
                    # eta = 2Ki,j -Ki,i - Kj,j
                    eta = 2.0 * dataMatrix[i, :] * dataMatrix[j, :].T - dataMatrix[i, :] * dataMatrix[i, :].T \
                                                                        - dataMatrix[j, :] * dataMatrix[j, :].T
                    if eta >= 0:
                        print ("eta >= 0")
                        continue
                        
                    # update alpha[j]
                    alphas[j] -= labelMat[j] * (Ei - Ej) / eta
                    alphas[j] = self.clipAlpha(alphas[j], H, L)
                    
                    # check the variation of alphas[j], this iteration will be skipped if the variation is too small
                    if (abs(alphas[j] - alphaJold) < 0.00001):
                        continue
                    
                    # Σaiyi = 0 --> yj(a_new[j] - a_old[j]) + yi(a_new[i] - a_old[i]) = 0
                    #           --> (a_new[i] - a_old[i]) = - (a_new[j] - a_old[j]) * yj / yi  两边同乘yi^2 (yi^2 = 1)
                    #           --> (a_new[i] - a_old[i]) = - (a_new[j] - a_old[j]) * yj * yi
                    #           --> a_new[i] = a_old[i] + (a_old[j] - a_new[j]) * yj * yi
                    alphas[i] += labelMat[j] * labelMat[i] * (alphaJold - alphas[j])
                    # after update alphaPairs, b will be updated.
                    b1 = b - Ei - labelMat[i]*(alphas[i]-alphaIold)*dataMatrix[i, :]*dataMatrix[i, :].T \
                                - labelMat[j]*(alphas[j]-alphaJold)*dataMatrix[i, :]*dataMatrix[j, :].T
                    b2 = b - Ej - labelMat[i]*(alphas[i]-alphaIold)*dataMatrix[i, :]*dataMatrix[j, :].T \
                                - labelMat[j]*(alphas[j]-alphaJold)*dataMatrix[j, :]*dataMatrix[j, :].T
                    
                    if (0 < alphas[i]) and (C > alphas[i]): # a[i] not in boundary
                        b = b1
                    elif (0 < alphas[j]) and (C > alphas[j]): # a[j] not in boundary
                        b = b2
                    else:
                        b = (b1 + b2) / 2
                    
                    # complete the alpha pair update
                    alphaPairsChanged += 1
                    print("iter: %d i:%d, pairs changed %d" % (iter, i, alphaPairsChanged))
                
                """
                内循环：对每个alpha[i]，随机选取另一个alpha[j]，update alpha pairs，遍历alphas。
                外循环：在一次内循环中，如果有alpha发生改变，则将iter重置为0.
                       否则iter加一。
                       直到连续运行maxiter次内循环后仍无任何alpha发生改变，说明此时已无法继续优化
                """
            if (alphaPairsChanged == 0):
                iter += 1
            else:
                iter = 0
            print ("iteration number: %d" % iter)
        return b, alphas
        
    def calcWs(self, alphas, dataArr, classLabels):
        """
        Desc:
            calcuate w with giving alphas
        Args:
            alphas -- lagrangian multiplier
            dataArr -- feature matrix, also X
            classLabels -- labels, also Y
        Returns:
            w -- w, the regression coeffients
        """
        X = np.mat(dataArr)
        labelMat = np.mat(classLabels).transpose()
        m, n = np.shape(X)
        w = np.zeros((n, 1))
        for i in range(m):
            # np.multiply : 对应元素相乘，当第一个为一个数时，为这个数乘以后面对应的所有元素
            w += np.multiply(alphas[i] * labelMat[i], X[i, :].T) # alpha[i] is a value, labelMat[i] is a value
        return w

    def plotfig_SVM(self, xMat, yMat, ws, b, alphas):
        """
        参考地址：
           http://blog.csdn.net/maoersong/article/details/24315633
           http://www.cnblogs.com/JustForCS/p/5283489.html
           http://blog.csdn.net/kkxgx/article/details/6951959
        """

        xMat = np.mat(xMat)
        yMat = np.mat(yMat)

        # b原来是矩阵，先转为数组类型后其数组大小为（1,1），所以后面加[0]，变为(1,)
        b = np.array(b)[0]
        fig = plt.figure()
        ax = fig.add_subplot(111)

        # 注意flatten的用法
        ax.scatter(xMat[:, 0].flatten().A[0], xMat[:, 1].flatten().A[0])

        # x最大值，最小值根据原数据集dataArr[:, 0]的大小而定
        x = np.arange(-1.0, 10.0, 0.1)

        # 根据x.w + b = 0 得到，其式子展开为w0.x1 + w1.x2 + b = 0, x2就是y值
        y = (-b-ws[0, 0]*x)/ws[1, 0]
        ax.plot(x, y)

        for i in range(np.shape(yMat[0, :])[1]):
            if yMat[0, i] > 0:
                ax.plot(xMat[i, 0], xMat[i, 1], 'cx')
            else:
                ax.plot(xMat[i, 0], xMat[i, 1], 'kp')

        # 找到支持向量，并在图中标红
        for i in range(100):
            if alphas[i] > 0.0:
                ax.plot(xMat[i, 0], xMat[i, 1], 'ro')
        plt.show()
                
```

```python
svm1 = SVM_simpleSMO()
# 获取特征和目标变量
dataArr, labelArr = svm1.loadDataSet(r'..\data\Ch06\testSet.txt')
b, alphas = svm1.smoSimple(dataArr, labelArr, 0.6, 0.001, 40)
```

```python
ws = svm1.calcWs(alphas, dataArr, labelArr)
svm1.plotfig_SVM(dataArr, labelArr, ws, b, alphas)
```
