```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class AdaBoost():
    def loadDataSet(self, filename):
        """
        Desc:
            load dataset from .txt file. 
            Note than the labels should equal to 1 or -1 but not 1 or 0.
        Args:
            filename -- .txt file path
        Returns:
            dataArr -- feature matrix
            labelArr -- labels
        """
        dataArr = []
        labelArr = []
        with open(filename) as fr:
            for line in fr.readlines():
                lineArr = []
                curLine = line.strip().split('\t')
                for i in range(len(curLine) - 1):
                    lineArr.append(float(curLine[i]))
                dataArr.append(lineArr)
                labelArr.append(float(curLine[-1]))
        return np.matrix(
            dataArr), labelArr  # convert list type dataArr into np.matrix type

    def stump_classify(self, data_mat, dimen, thresh_val, thresh_ineq):
        """
        Desc:
            sort the giving feature value and use dichotomization(二分法) to predict the results
        Args:
            data_mat -- feature matrix, np.mat type
            dimen -- the feature index used to split dataset
            thresh_val --threshold for classification
            thresh_ineq -- indicate that we change the values on the left or on the right
        Returns:
            ret_array -- the classification results
        
        Note:
            Q: Why we need to add a thresh_ineq?
            A: Because we don't we should change left values or right values
                在根据threshold将dataset划分成两部分后，我们并不知道小于threshold的那部分到底
                应该为-1还是为1，因此在本程序中，同时计算了小于threshold为1、为-1的两种情况，
                并根据最终的权重error判断到底应该为1还是为-1，thresh_ineq也就是为了记录这件事最终结果。
        """
        ret_array = np.ones((np.shape(data_mat)[0], 1))  # (m,1)

        # 其实就是互相颠倒1和-1，然后看weighted error哪种小
        if thresh_ineq == 'lt':
            ret_array[data_mat[:, dimen] <= thresh_val] = -1
        else:
            ret_array[data_mat[:, dimen] > thresh_val] = -1
        return ret_array

    def build_stump(self, data_arr, class_labels, D):
        """
        Desc:
            train a decision stump with a giving D vector
        Args:
            data_arr -- feature matrix
            class_labels -- sample labels
            D -- sample weight vector, (m,1)
        Returns:
            best_stump -- best decision stump model, dict type, contain 
                                1. the best feature index
                                2. the best threshold
                                3. inequal indicator
            min_err -- the minimum weighted error
            best_class_est -- the classification results predicted by the best decision stump
                            
        """
        data_mat = np.mat(data_arr)
        label_mat = np.mat(class_labels).T
        
        m, n = np.shape(data_mat)
        num_steps = 10.0
        best_stump = {}
        best_class_est = np.mat(np.zeros((m, 1)))
        min_err = np.inf
        
        for i in range(n):  # traverse all features
            # gain the increasing step size for threshold
            range_min = data_mat[:, i].min()
            range_max = data_mat[:, i].max()
            step_size = (range_max - range_min) / num_steps
            for j in range(-1, int(num_steps), +1):
                # calculate threshold
                # we use this threshold to split dataset and predict the results
                # 单层决策树, 根据最优特征的最优阈值进行二分类，使得权重error最小
                # 权重error为分类结果对错乘以权重vector D
                thresh_val = range_min + float(j) * step_size
                for inequal in ['lt', 'gt']:
                    # predict label for each sample with giving feature index and threshold
                    predicted_vals = self.stump_classify(
                        data_mat, i, thresh_val, inequal)

                    # count the error classified sample number
                    err_arr = np.mat(np.ones((m, 1)))
                    err_arr[predicted_vals == label_mat] = 0

                    # calculate the weighted error, D:(m,1), err_arr:(m,1)
                    weighted_err = D.T * err_arr

                    # find the best feature and threshold which can minimize the weighted error
                    if weighted_err < min_err:
                        min_err = weighted_err
                        best_class_est = predicted_vals.copy()
                        best_stump['dim'] = i
                        best_stump['thresh'] = thresh_val
                        best_stump['ineq'] = inequal

        return best_stump, min_err, best_class_est

    def adaBoostTrainDS(self, data_arr, class_labels, num_it=40):
        """
        Desc:
            train a series of weak classifiers and use them to predict labels.
            Each weak classifier is a decision stump(单层决策树).
            所谓多个弱分类器指的是使用同一个数据集上训练多次，每训练一次生成一个弱分类器，
            下一个弱分类器的D与alpha通过本次的分类结果计算获得。
            Ending condition: 
                1. the generated weak classifier number reaches to the giving number
                2. the error rate = 0
        Args:
            data_arr -- feature matrix
            class_labels -- sample labels
            nim_it -- iteration time, also weak classifier number
        Returns:
            weak_class_arr -- the collection of weak classifiers
            agg_class_est -- the final classification results
        """
        weakClassArr = []  # to store the weak classifiers
        m = np.shape(data_arr)[0]

        # 默认向量为列向量
        # Each sample should have a corresponding weight coeffient D, so D is a vector (m,1)
        D = np.mat(np.ones((m, 1)) / m)  # Intialization: uniform D
        agg_class_est = np.mat(np.zeros((m, 1)))
        for i in range(num_it):
            # build decision tree model
            best_stump, error, class_est = self.build_stump(
                data_arr, class_labels, D)

            # update alpha, each classifier has a alpha
            # 每训练一次，我们只得到一个弱分类器，因此在循环的每一步中，我们只得到1个alpha
            # add max() to avoid zero in denominator
            alpha = float(0.5 * np.log((1.0 - error) / max(error, 1e-16)))
            best_stump['alpha'] = alpha
            weakClassArr.append(best_stump)  # store stump parameters

            # 分类正确：乘积为1，不会影响结果，-1主要是下面求e的-alpha次方
            # 分类错误：乘积为 -1，结果会受影响，所以也乘以 -1
            # class_est为分类器预测结果，为(m,1)的vector
            # class_labels为实际label,初始为list，通过np.mat()并转置使其保持与class_est的shape一直
            # np.mat([0]*10).shape，使用np.mat()处理List，默认得到的是行向量
            expon = np.multiply(-1 * alpha * np.mat(class_labels).T, class_est)
            # update D, D is a vector (m,1), np.mat type
            D = np.multiply(D, np.exp(expon))
            D = D / D.sum()

            # calculate the final classification results
            agg_class_est += alpha * class_est

            # 最后的分类结果根据预测值的符号判断.
            # np.sign(agg_class_est) can get the final classification results.
            # 因为是!=，因此在预测正确的时候，括号内乘积为0，预测错误时，结果为1
            agg_errors = np.multiply(
                (np.sign(agg_class_est) != np.mat(class_labels).T),
                np.ones((m, 1)))
            error_rate = agg_errors.sum() / m  # calcuate the error rate.
            if error_rate == 0.0:
                break
        return weakClassArr, agg_class_est
    
    def plot_roc(self, pred_strengths, class_labels):
        """
        Desc:
            plot the ROC and calculate AUC
            Q: How to plot ROC?
            A: 
            1. sort the classification results and get the corresponding index
            2. start from position (1,1), check the actual label value 
                                    (label的顺序是根据上一步分类结果排序的)
                If the value = 1, then move to (1, 1 - y_step) 即向下移动一个y步长
                If the value = -1, then move to (1 - x_step, 1) 即向左移动一个x补偿
                The step size is calculated by the number of positive samples and negetive samples
                x轴代表negetive sample, y轴代表Positive sample
            3. now we start from the second position, then repeat the last step, until all the results
                are used. We can easily know that the finaly position must be (0, 0)
            
        Args:
            pred_strengths -- the weighted classification results (m,1)
            class_labels -- actual labels (m,1)
        Returns:
            None
            
        """
        y_sum = 0.0
        # sum the positive samples
        num_pos_class = np.sum(np.array(class_labels) == 1.0)
        y_step = 1 / float(num_pos_class)
        
        x_step = 1 / float(len(class_labels) - num_pos_class)
        
        # np.argsort() 返回将数组从小到大排列的索引
        sorted_indicies = pred_strengths.reshape(-1).argsort()
        
        fig = plt.figure()
        fig.clf()
        ax = plt.subplot(111)
        cur = (1.0, 1.0) # initial position
        
        for index in sorted_indicies.tolist()[0]:
            if class_labels[index] == 1.0:
                del_x = 0
                del_y = y_step
            else:
                del_x = x_step
                del_y = 0
                y_sum += cur[1] # used to calculate AUC
            ax.plot([cur[0], cur[0] - del_x], [cur[1], cur[1] - del_y], c = 'b')
            cur = (cur[0] - del_x, cur[1] - del_y) # change the position
        
        # plot the diagonal line with dash line
        ax.plot([0, 1], [0, 1], 'b--')
        plt.xlabel('False positive rate')
        plt.ylabel('True positive rate')
        plt.title('ROC curve for AdaBoost horse colic detection system')
        # 设置画图的范围区间 (x1, x2, y1, y2)
        ax.axis([0, 1, 0, 1])
        plt.show()
        
        # calculate AUC
        print("the Area Under the Curve is: ", y_sum * x_step)
        

    def ada_classify(self, data_to_class, classifier_arr):
        """
        Desc:
            predict labels for the giving dataset by using the series of weak classifiers
        Args:
            data_to_class -- input dataset
            classifier_arr -- the series of weak classifiers, list-type
                                every element is a dictionary (best_stump)
        Return:
            the classification result (1 or -1)
        """
        data_mat = np.mat(data_to_class)
        m = np.shape(data_mat)[0]
        agg_class_est = np.mat(np.zeros((m, 1)))
        for i in range(len(classifier_arr)):
            # predict labels for input dataset with each weak classifier
            class_est = self.stump_classify(data_mat,
                                            classifier_arr[i]['dim'],
                                            classifier_arr[i]['thresh'],
                                            classifier_arr[i]['ineq'])
            # calculate the weighted classification result
            # class_est is a vector, (m, 1), so agg_class_est is also a vector (m, 1)
            agg_class_est += classifier_arr[i]['alpha'] * class_est
            print (agg_class_est)
        return np.sign(agg_class_est)
```
